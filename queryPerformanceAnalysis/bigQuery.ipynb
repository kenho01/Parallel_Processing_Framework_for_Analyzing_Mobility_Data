```python
from google.cloud import bigquery

client = bigquery.Client()
query = """
SELECT d.DRIVER_ID, AVG(duration_s) AS average_trip_duration, AVG(distance_km) AS average_trip_distance
FROM `mobilitydata.fact_trips` f
LEFT JOIN `mobilitydata.dim_vehicle` v ON f.VEHICLE_ID = v.VEHICLE_ID
LEFT JOIN `mobilitydata.dim_driver` d ON f.DRIVER_ID = d.DRIVER_ID
LEFT JOIN `mobilitydata.dim_pickup_time` pt ON f.DT_START_ID = pt.PICKUP_TIME_ID
LEFT JOIN `mobilitydata.dim_dropoff_time` do ON f.DT_END_ID = do.DROPOFF_TIME_ID
LEFT JOIN `mobilitydata.dim_pickup_location`pu ON f.PICKUP_LOCATION_ID = pu.PICKUP_LOCATION_ID
LEFT JOIN `mobilitydata.dim_dropoff_location` dl ON f.DROPOFF_LOCATION_ID = dl.DROPOFF_LOCATION_ID
GROUP BY d.DRIVER_ID

"""
results = client.query(query)
df = results.to_dataframe()
display(df)
```
